```python
import numpy as np

# Step 1: User input for number of products
n = int(input("Enter number of products: "))

quantities = []
prices = []

# Step 2: Collect user input
for i in range(n):
    print("\nProduct", i + 1)

    qty = float(input("Enter quantity: "))
    price = float(input("Enter price: "))

    quantities.append(qty)
    prices.append(price)

# Step 3: Convert lists to NumPy arrays
qty_array = np.array(quantities)
price_array = np.array(prices)

# Step 4: Broadcasting (vectorized computation)
revenue = qty_array * price_array

# Step 5: Display results
print("\n--- Sales Report ---")
print("Quantities:", qty_array)
print("Unit Prices:", price_array)
print("Revenue:", revenue)

# Step 6: Statistical Analysis
print("\n--- Statistical Analysis ---")
print("Total Revenue:", np.sum(revenue))
print("Average Revenue:", np.mean(revenue))
print("Maximum Revenue:", np.max(revenue))
print("Minimum Revenue:", np.min(revenue))
print("Standard Deviation:", np.std(revenue))
print("Variance:", np.var(revenue))
```
